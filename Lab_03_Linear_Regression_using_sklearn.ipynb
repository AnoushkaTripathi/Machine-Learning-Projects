```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/content/student_scores.csv")
data.head()
```

```python
data.describe()
```

```python
X=data.drop('Scores',axis=1)
X.head()
```

```python
y=data['Scores']
y.head()
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest =train_test_split(X,y,test_size=0.2,random_state=1)
print(xTrain.head())
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest =train_test_split(X,y,test_size=0.2,random_state=1)
print(xTrain.head())
print(xTest.head())
print(yTrain.head())
print(xTrain.head())
```

```python
xTrain.shape,xTest.shape,yTrain.shape,yTest.shape
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(xTrain,yTrain)
```

```python
print(LR.intercept_)
print(LR.coef_)
```

```python
y_pred=LR.predict(xTest)
df = pd.DataFrame({'Actual':yTest,'Predicted':y_pred})
df
```

```python
from sklearn import metrics
print('Mean Absolute error:',metrics.mean_absolute_error(yTest,y_pred))
print('Mean Squared error:',metrics.mean_squared_error(yTest,y_pred))
print('Root Mean Absolute error:',np.sqrt(metrics.mean_absolute_error(yTest,y_pred)))
print('Mean Percentage error:',np.mean((yTest-y_pred)/yTest)*100)
print('R2 Score:',metrics.r2_score(yTest,y_pred))

```
